```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
db = pd.read_csv("cars.csv",encoding='latin1')
```

```python
db.head()
```

```python
#Fetching number of rows and column
print ("(row,column)")
db.shape
```

```python
db.describe(include = 'all')
```

```python
relevant_columns = db[['mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']]
# Calculate the mean for each column
mean_values = relevant_columns.mean()
print("Mean:\n", mean_values)
print()
# Calculate the median for each column
median_values = relevant_columns.median()
print("Median:\n", median_values)
print()
# Calculate the standard deviation for each column
std_values = relevant_columns.std()
print("Standard Deviation:\n", std_values)
```

```python
import seaborn as sns
for i in relevant_columns:
    db[i] = db[i].fillna(db[i].mean())
    plt.figure(figsize=(4,2))
    sns.boxplot(db[i],color = "yellow")
    plt.title('Interpretation of 25%ile and 75%ile')
    plt.show()
    print("25% = ",np.percentile(db[i], 25))
    print("75% = ",np.percentile(db[i], 75))
```

```python
# Setting up the plotting style
sns.set(style="whitegrid")

# Figure 1: Distribution of MPG (Miles per Gallon)
plt.figure(figsize=(8, 5))
sns.histplot(db['mpg'], kde=True, bins=10, color='skyblue')
plt.title("Distribution of Miles per Gallon (MPG)", fontsize=14)
plt.xlabel("MPG", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.show()

# Figure 2: Scatter plot of Horsepower (hp) vs. Weight (wt)
plt.figure(figsize=(8, 5))
sns.scatterplot(x=db['wt'], y=db['hp'], hue=db['cyl'], palette="viridis", s=100)
plt.title("Horsepower vs. Weight", fontsize=14)
plt.xlabel("Weight (1000 lbs)", fontsize=12)
plt.ylabel("Horsepower", fontsize=12)
plt.legend(title="Cylinders")
plt.show()

# Figure 3: Boxplot of MPG by Cylinder Count
plt.figure(figsize=(8, 5))
sns.boxplot(x=db['cyl'], y=db['mpg'])
plt.title("MPG by Cylinder Count", fontsize=14)
plt.xlabel("Number of Cylinders", fontsize=12)
plt.ylabel("Miles per Gallon (MPG)", fontsize=12)
plt.show()

# Figure 4: Pairplot of selected features
selected_features = ['mpg', 'hp', 'wt', 'disp']
sns.pairplot(db[selected_features], diag_kind="kde", corner=True,)
plt.suptitle("Pairwise Relationships Between Selected Features", y=1.02, fontsize=14)
plt.show()
```
